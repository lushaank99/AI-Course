```python
from collections import deque
import copy
def find_bfs(input_seq, output_seq, graph):
    if input_seq == output_seq:
        print(input_seq)
        return
    queue = deque([input_seq])
    parent = []
    parent.append([input_seq])
    while queue:
        currstr = queue.popleft()
        num = currstr.index(0)
        print(currstr, "\n")
        for nei in graph[num]:
            tempstr = copy.deepcopy(currstr)
            tempstr[nei], tempstr[num] = tempstr[num], tempstr[nei]
            if tempstr == output_seq:
                print(tempstr)
                return
            if (tempstr not in parent) and (tempstr not in queue):
                parent.append(tempstr)
                queue.append(tempstr)
    print("No solution found")
    return
```

```python
def find_dfs(input_seq, output_seq, graph):
    if input_seq == output_seq:
        print(input_seq)
        return
    queue = deque([input_seq])
    parent = []
    parent.append([input_seq])
    while queue:
        currstr = queue.pop()
        num = currstr.index(0)
        print(currstr, "\n")
        for nei in graph[num]:
            tempstr = copy.deepcopy(currstr)
            tempstr[nei], tempstr[num] = tempstr[num], tempstr[nei]
            if tempstr == output_seq:
                print(tempstr)
                return
            if (tempstr not in parent) and (tempstr not in queue):
                parent.append(tempstr)
                queue.append(tempstr)
    print("No solution found")
    return
```

```python
def cost(state):
    cost_list = [[0, 1, 2, 1, 2, 3, 2, 3, 4], [1, 0, 1, 2, 1, 2, 3, 2, 3], [2, 1, 0, 3, 2, 1, 4, 3, 2], [1, 2, 3, 0, 1, 2, 1, 2, 3], [2, 1, 2, 1, 0, 1, 2, 1, 2], [3, 2, 1, 2, 1, 0, 3, 2, 1], [2, 3, 4, 1, 2, 3, 0, 1, 2], [3, 2, 3, 2, 1, 2, 1, 0, 1], [4, 3, 2, 3, 2, 1, 2, 1, 0]]
    value = 0
    for item in state:
        ind = state.index(item)
        value += cost_list[item][ind]
    return value
    
def heuristic1(next_state):
    min_value = 100
    for state in next_state:
        value = cost(state)
        if value < min_value:
            min_value = value
            fin = state
    return(fin)
```

```python
def find_greedy(input_seq, output_seq, graph):
    if input_seq == output_seq:
        print(input_seq)
        return
    visited = []
    visited.append(input_seq)
    print(input_seq)
    while len(visited) < 1000000:
        next_state = []
        tempstr = copy.deepcopy(input_seq)
        num = tempstr.index(0)
        for nei in graph[num]:
            tempstr[nei], tempstr[num] = tempstr[num], tempstr[nei]
            if tempstr == output_seq:
                print(tempstr)
                return
            if (tempstr not in visited):
                next_state.append(tempstr)
        next_heu = heuristic1(next_state)
        if next_heu in visited:
            break
        input_seq = next_heu
        visited.append(next_heu)
        print(next_heu)
    print("No solution found")
    return
```

```python
def heuristic2(next_state, length):
    min_value = 100
    for state in next_state:
        value = cost(state)
        if (value + length) < min_value:
            min_value = value + length
            fin = state
    return(fin)
```

```python
def find_astar(input_seq, output_seq, graph):
    if input_seq == output_seq:
        print(input_seq)
        return
    visited = []
    i = 0
    visited.append([input_seq, i])
    print(input_seq)
    while len(visited) < 1000000:
        next_state = []
        tempstr = copy.deepcopy(input_seq)
        num = tempstr.index(0)
        i += 1
        for nei in graph[num]:
            tempstr[nei], tempstr[num] = tempstr[num], tempstr[nei]
            if tempstr == output_seq:
                print(tempstr)
                return
            if (tempstr not in visited):
                next_state.append(tempstr)
        next_heu = heuristic2(next_state, i)
        if next_heu in visited:
            break
        input_seq = next_heu
        visited.append(next_heu)
        print(next_heu)
    print("No solution found")
    return
```
